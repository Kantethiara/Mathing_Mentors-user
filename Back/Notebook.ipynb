```python
from sentence_transformers import SentenceTransformer, util
import pandas as pd
```

```python
model_embed = SentenceTransformer("all-MiniLM-L6-v2")

def cosine_similarity_np(a, b):
    a = np.array(a)
    b = np.array(b)
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

def text_similarity(a, b):
    if not isinstance(a, str) or not isinstance(b, str):
        return 0.0
    emb_a = model_embed.encode(a, convert_to_numpy=True).flatten()
    emb_b = model_embed.encode(b, convert_to_numpy=True).flatten()
    return float(np.dot(emb_a, emb_b) / (np.linalg.norm(emb_a) * np.linalg.norm(emb_b)))


def compute_pair_features(student, mentor):
    return {"student_id": student['student_id'],
        "mentor_id": mentor['mentor_id'],
        "competence_sim": text_similarity(student["Quelle formation serait idéale pour toi ?"], mentor["Liste les compétences clés que tu peux transmettre à des apprenants dans ce domaine ou métier"]),
        "objectif_sim": text_similarity(student["Quel est ton principal objectif professionnel ?"], mentor['Domaine principal d\'expertise']),
        "carrer_sim": text_similarity(student["Quelle est ton métier et ta situation professionnelle actuelle ?"], mentor['Quel est ton métier et ta situation professionnelle actuelle ?']),
        "expertise_sim": text_similarity(student["Quelle formation serait idéale pour toi ?"],mentor['Domaine principal d\'expertise']),
    }
```

```python
compute_pair_features(df_student, df_mentor)
```

```python
ranking_data = []

for _, student in df_student.iterrows():
    for _, mentor in df_mentor.iterrows():
        features = compute_pair_features(student, mentor)
        features["score"] = (
            0.3 * features["competence_sim"] +
            0.2 * features["objectif_sim"] +
            0.2 * features["carrer_sim"] +
            0.3 * features["expertise_sim"]
        )
        ranking_data.append(features)

ranking_df = pd.DataFrame(ranking_data)
```

```python
def score_to_relevance(score):
    if score >= 0.5:
        return 3
    elif score >= 0.25:
        return 2
    else:
        return 0
ranking_df["relevance"] = ranking_df["score"].apply(score_to_relevance)
```

```python
import lightgbm as lgb

features = ['competence_sim', 'objectif_sim','carrer_sim', 'expertise_sim']
X = ranking_df[features]
y = ranking_df["relevance"]

group = ranking_df.groupby("student_id").size().tolist()
# print(group)
lgb_data = lgb.Dataset(X, label=y, group=group)

params = {
    "objective": "lambdarank",
    "metric": "ndcg",
    "ndcg_eval_at": [1, 3, 5],
    "learning_rate": 0.1,
}

model = lgb.train(params, lgb_data, num_boost_round=10)
```

```python
model.predict(ranking_df[features].head(20))
```

```python
def recommend_mentors_real_time(new_student, df_mentor, trained_model,predictors, top_n=5):
    feature_list = []
    meta = []

    for _, mentor in df_mentor.iterrows():
        features = compute_pair_features(new_student, mentor)
        feature_list.append([
            features["competence_sim"],
            features["objectif_sim"],
            features["carrer_sim"],
            features["expertise_sim"],
        ])
        meta.append({
            "mentor_id": mentor["mentor_id"],
            "mentor_email": mentor.get("Adresse e-mail", ""),
        })

    feature_df = pd.DataFrame(feature_list, columns=predictors)

    scores = trained_model.predict(feature_df)
    result_df = pd.DataFrame(meta)
    result_df["score"] = scores

    return result_df.sort_values(by="score", ascending=False).head(top_n)
```
